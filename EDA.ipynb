identifying IQR

```python
## Numpy to identify the IQR
q1 = np.percentile(required_column['Radius (mean)'], 25)
q3 = np.percentile(required_column['Radius (mean)'], 75)
iqr = q3 - q1
iqr
```

```python
### Pandas to identify the IQR
q1 = required_column['Radius (mean)'].quantile(0.25)
q3 = required_column['Radius (mean)'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print(lower_bound)
print(upper_bound)
```

```python
final_Radius = required_column[(required_column['Radius (mean)'] >= lower_bound) & (required_column['Radius (mean)'] <= upper_bound)]
final_Radius.describe()
```

#### Visual REpresentation of Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

np.random.seed(42)
x = np.random.normal(50,10,100)
y = np.random.normal(0,5,100)

df = pd.DataFrame({'x': x, 'y': y})

df.iloc[5] = [50,200] ## outlier for y
df.iloc[25] = [90,20] ## outlier for x

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

is_inside_bounds = (df >= (Q1 - 1.5 * IQR)) & (df <= (Q3 + 1.5 * IQR))
df['is_outlier'] = ~is_inside_bounds.all(axis=1)
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
sns.scatterplot(data=df, x='x', y='y', hue='is_outlier', ax=axes[0]), axes[0].set_title('Scatter Plot with Outliers')

df_no_outliers = df[~df['is_outlier']]
sns.scatterplot(data=df_no_outliers, x='x', y='y', ax=axes[1]), axes[1].set_title('Scatter Plot without Outliers')
plt.tight_layout()
plt.show()
```

identifying IQR

```python
## Numpy to identify the IQR
q1 = np.percentile(required_column['Radius (mean)'], 25)
q3 = np.percentile(required_column['Radius (mean)'], 75)
iqr = q3 - q1
iqr
```

```python
### Pandas to identify the IQR
q1 = required_column['Radius (mean)'].quantile(0.25)
q3 = required_column['Radius (mean)'].quantile(0.75)
iqr = q3 - q1
iqr
```

```python
lower_bound = q1 - (1.5 * iqr)
upper_bound = q3 + (1.5 * iqr)
print(lower_bound)
print(upper_bound)
```

```python
final_Radius = required_column[(required_column['Radius (mean)'] >= lower_bound) & (required_column['Radius (mean)'] <= upper_bound)]
final_Radius.describe()
```

## Sample Data 

```python
data = {
    'name': ['Xavier', 'Ann', 'Jana', 'Yi', 'Robin', 'Amal', 'Nori'],
    'city': ['Mexico City', 'Toronto', 'Prague', 'Shanghai', 'Manchester', 'Cairo', 'Osaka'],
    'age': [41, 28, 33, 34, 38, 31, 37],
    'py-score': [88.0, 79.0, 81.0, 80.0, 68.0, 61.0, 84.0]
 }

row_labels = [101, 102, 103, 104, 105, 106, 107]
```

creating dataframe 

```python
df = pd.DataFrame(data, index=row_labels)
df
```

```python
df['java-score'] = [92.0, 85.0, 78.0, 88.0, 72.0, 65.0, 90.0]
df
```

```python
df['final-score'] = (df['py-score'] + df['java-score']) / 2 
df
```

```python
df['last-name'] = ['', 'Smith', '', 'Chen', '', 'Ali', 'Tanaka']
df
```

Apply function applies lambda to whole columns

```python
df['full-name'] = df.apply(lambda x: x['name'] + ' ' + x['last-name'], axis=1)
df
```

filtering 

```python
filter = df['age'] > 30
df[filter]
```

## Working with time-series data

```python
temp_c = [ 8.0,  7.1,  6.8,  6.4,  6.0,  5.4,  4.8,  5.0, 9.1, 12.8, 15.3, 19.1, 21.2, 22.1, 22.4, 23.1, 21.0, 17.9, 15.5, 14.4, 11.9, 11.0, 10.2,  9.1]
```

```python
dt = pd.date_range(start='2019-10-27 00:00:00.0', periods=24,freq='h')
dt
```

```python
temperature = pd.DataFrame(temp_c, index=dt, columns=['Temperature'])
temperature
```
